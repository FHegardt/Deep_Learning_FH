```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
fashion_mnist = tf.keras.datasets.fashion_mnist.load_data()
```

```python
def display_images(nrows = 2, ncols = 5, figsize =(16,4)):

    fig, axes = plt.subplots(nrows, ncols, figsize = figsize)

    for i, ax in enumerate(axes.flatten()):
        ax.imshow(fashion_mnist[0][0][i], cmap = "gray")
        ax.axis("off")
    
    fig.subplots_adjust(wspace=0, hspace=.1, bottom=0)

display_images()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
def display_images_scaled(nrows = 2, ncols = 5, figsize =(16,4)):

    fig, axes = plt.subplots(nrows, ncols, figsize = figsize)

    for i, ax in enumerate(axes.flatten()):
        scaled = scaler.fit_transform(fashion_mnist[0][0][i])
        ax.imshow(scaled, cmap = "gray")
        ax.axis("off")
    
    fig.subplots_adjust(wspace=0, hspace=.1, bottom=0)

display_images_scaled()
```
